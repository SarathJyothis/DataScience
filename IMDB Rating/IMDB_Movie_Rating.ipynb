```python
#Classify movies with Very Good-2, Good-1 and Bad-0 based on Average Rating
data_movies1.head()
```

```python

def f(row):
    if row['averageRating'] <= 5:
        val = 0
    elif row['averageRating'] > 5 and row['averageRating'] <= 7.5:
        val = 1
    else:
        val = 2
    return val
```

```python
data_movies1['AvgRating'] = data_movies1.apply(f, axis=1)
data_movies1.head()
```

```python
#Replace '/N' with 0
```

```python
data_movies1["averageRating"] = data_movies1["averageRating"].replace("\\N", 0)
data_movies1["isAdult"] = data_movies1["isAdult"].replace("\\N", 0)
data_movies1["runtimeMinutes"] = data_movies1["runtimeMinutes"].replace("NA", 0)
data_movies1["numVotes"] = data_movies1["numVotes"].replace("\\N", 0)
data_movies1["ordering"] = data_movies1["ordering"].replace("\\N", 0)
data_movies1["birthYear"] = data_movies1["birthYear"].replace("\\N", 0)
data_movies1["title_type"] = data_movies1["title_type"].replace("\\N", 0)
data_movies1["genres_type"] = data_movies1["genres_type"].replace("\\N", 0)
data_movies1["Writers"] = data_movies1["Writers"].replace("\\N", 0)
data_movies1["Category"] = data_movies1["Category"].replace("\\N", 0)
data_movies1["Musical"] = data_movies1["Musical"].replace("\\N", 0)
data_movies1["Sport"] = data_movies1["Sport"].replace("\\N", 0)
data_movies1["Romance"] = data_movies1["Romance"].replace("\\N", 0)
data_movies1["Short"] = data_movies1["Short"].replace("\\N", 0)
data_movies1["Drama"] = data_movies1["Drama"].replace("\\N", 0)
data_movies1["Music"] = data_movies1["Music"].replace("\\N", 0)
data_movies1["Crime"] = data_movies1["Crime"].replace("\\N", 0)
data_movies1["Action"] = data_movies1["Action"].replace("\\N", 0)
data_movies1["Comedy"] = data_movies1["Comedy"].replace("\\N", 0)
data_movies1["History"] = data_movies1["History"].replace("\\N", 0)

data_movies1["Film-Noir"] = data_movies1["Film-Noir"].replace("\\N", 0)
data_movies1["News"] = data_movies1["News"].replace("\\N", 0)
```
